##### Scatter plot visualization based on data from June -> Dec '17

```python
import numpy             as np
import pandas            as pd
import seaborn           as sns
import matplotlib.pyplot as plt
%matplotlib inline


from sklearn.cluster       import KMeans, MiniBatchKMeans
from sklearn.preprocessing import scale
```

```python
DF_Parcel = pd.read_csv("Data/Parcel Pick Data - 20170601 to 20171211.csv")
DF_Parcel.head()
```

```python
# Drop all of the columns with values which are not numbers. Reduce data size to make it easier to work with
# Can also perform k-means on categorical data, later. See here:
# https://datascience.stackexchange.com/questions/22/k-means-clustering-for-mixed-numeric-and-categorical-data
onlyNumDF = DF_Parcel
onlyNumDF.drop(['UserId', 'Aisle', 'Bin', 'ZoneCode', 'PickDate'], axis=1, inplace=True)
onlyNumDF.iloc[:2001] 
```

```python
onlyNumDF['Bay'].max()
```

```python
onlyNumDF.info()
```

```python
# NOTE: dropna does not change data set. Weird.
# onlyNumDF.dropna()

# Fill missing values with averages, otherwise we can't run the ML model
onlyNumDF.fillna(onlyNumDF.mean(), inplace=True)
```

```python
onlyNumDF.info()
```

```python
onlyNumDF.head()
```

# Build the model

```python
model = KMeans(n_clusters = 5)
```

```python
model = model.fit(scale(onlyNumDF))
```

```python
print model.labels_
```

# Convert to matrix (array) so that matplotlib can display scatter plot

```python
onlyNumDF_toMatrix = onlyNumDF.as_matrix()
```

```python
onlyNumDF_toMatrix
```

# Scatter plots

```python
plt.figure(figsize = (8,6))
# Assign colors to label with an arbitrary value
plt.scatter(onlyNumDF_toMatrix[:,0], onlyNumDF_toMatrix[:,1], c = model.labels_.astype(np.float))
plt.xlabel('Bay'  )
plt.ylabel('Level')
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.scatter(onlyNumDF_toMatrix[:,1], onlyNumDF_toMatrix[:,2], c = model.labels_.astype(np.float))
plt.xlabel('Level'  )
plt.ylabel('Bay Pos')
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.scatter(onlyNumDF_toMatrix[:,2], onlyNumDF_toMatrix[:,3], c = model.labels_.astype(np.float))
plt.xlabel('Bay Pos')
plt.ylabel('Qty'    )
plt.show()

# TODO: Remove 5k outlier
```
